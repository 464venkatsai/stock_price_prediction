```python
# since the trades column has the missing values filling the missing values using the simpleimputer and standardizing the values
# making a pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
mypipeline = Pipeline([('imputer',SimpleImputer(strategy='mean')),('std_mean',StandardScaler())])
stock_data.iloc[:,3:] =mypipeline.fit_transform(stock_data.iloc[:,3:])
```

```python
# spliting the data in the 8:2 data by using str
from sklearn.model_selection import train_test_split
train_data,test_data = train_test_split(stock_data,random_state=40,test_size=0.2)
# splited = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=50)
# splited
```

```python
# preparing a model 
x_train = train_data.iloc[:,3:].drop('Close',axis=1).copy()
y_train =train_data['Close'].copy()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model_predictions=model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_train,model_predictions)
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.model_selection import cross_val_score
scores  = cross_val_score(model,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
rmse_after_cross_val=np.sqrt(-scores)
rmse_after_cross_val.mean()
```

```python
# testing the data
x_test,y_test = test_data.iloc[:,3:].drop('Close',axis=1).copy(),test_data['Close'].copy()
```

```python
final_prediction=model.predict(x_test)
final_mse = mean_squared_error(y_test,final_prediction)
final_rmse = np.sqrt(final_mse)
final_rmse
```

```python
scores  = cross_val_score(model,x_test,y_test,scoring='neg_mean_squared_error',cv=10)
rmse_after_cross_val=np.sqrt(-scores)
rmse_after_cross_val.mean()
```
